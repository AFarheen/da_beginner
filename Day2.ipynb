# Flip-coin game

```python
import random
guess = ''
while guess not in ('heads','tails'):
    print ('Guess the coin toss! Enter heads or tails :')
    guess = input()
toss = random.randint(0,1) # 0 is tails and 1 is heads
if toss == guess :
    print ('you got it!')
else:
    print ('Nope! Guess again!')
    guesss = input()
    if toss == guesss :
        print ('you got it!')
    else:
        print ('Nope, you are really bad at this game.')

    
```
