# Functions

```python
def convert_to_fahrenheit(celsiusTemp) :
    prod = celsiusTemp * 18
    final = prod + 32
    return final
```

```python
convert_to_fahrenheit(35)
```

# Tuples


```python
tuple = ('abcd', 786, 2.23, 'john', 70.2)
tinytuple = (123, 'john')
print (tuple) #prints complete tuple
print (tuple[0]) # prints first element of the tuple
print (tuple[1:3]) # prints elements starting from 2 till 3rd
print (tuple[2:]) # prints elements starting from 3rd
print (tinytuple * 2) # prints tinytuple 2 times
print (tuple + tinytuple) # prints concatenated tuples
```

```python
print (tuple)
```

```python
tuple[2] = 1000 #invalid syntax as we cannot edit/update tuple values once defined - throws type error
```
