```python
import pandas as pd
```

# Create a series object from a Python List

```python
ice_cream = ["choclate", "vanilla", "strawberry", "Rum raisin"]
```

```python
pd.Series(ice_cream)
```

```python
lottery = [4,8,16,24,32]

pd.Series(lottery)
```

```python
registrations = [True, False, False, True, False]

pd.Series(registrations)
```

# Create a series object from a Dictionary

```python
webster = {"Tiger" : "An Animal",
           "Banana" : "A Fruit",
           "Blue" : " A color"}

pd.Series(webster)
```
